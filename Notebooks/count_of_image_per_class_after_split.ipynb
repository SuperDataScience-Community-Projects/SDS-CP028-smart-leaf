```python
import os
from collections import defaultdict

base_dir = "C:/Users/OWNER/Downloads/SDS-CP028-smart-leaf_Project/SDS-CP028-smart-leaf/submissions/team-members/Samsudeen/Data/Train_Val_Test_Data"

split_counts = defaultdict(dict)

for split in ['train', 'val', 'test']:
    split_path = os.path.join(base_dir, split)
    print(f"\nüìÅ {split.upper()}:")

    for class_name in sorted(os.listdir(split_path)):
        class_path = os.path.join(split_path, class_name)
        if not os.path.isdir(class_path):
            continue

        count = len([
            f for f in os.listdir(class_path)
            if os.path.isfile(os.path.join(class_path, f))
        ])

        split_counts[split][class_name] = count
        print(f"  {class_name}: {count} images")

# Optional: Summary
print("\nüìä Summary Totals:")
for split in ['train', 'val', 'test']:
    total = sum(split_counts[split].values())
    print(f"  {split}: {total} images")
```
